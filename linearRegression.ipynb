```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures 

from sklearn.metrics import mean_squared_error
from sklearn import metrics

df = pd.read_csv('kc_house_data.csv') 
df['price'].map(int)
df.head()
```

```python
plt.plot(df['sqft_living15'] , df['sqft_living'] , 'ro')
plt.show()
# sqft_living with price gives a a little approch of a liner function 
# grade gives a shape of a poly fnction with price 
```

```python
#linear regression
x=df["sqft_living"].values.reshape(-1,1)
y=df["price"].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.35,random_state=40)
print(x_train.shape)
print(y_train.shape)

model=LinearRegression()  
model.fit(x_train,y_train)  
predicted=model.predict(x_test) #testing our modelâ€™s performance
print("MSE", mean_squared_error(y_test,predicted))
print("R squared", metrics.r2_score(y_test,predicted))
plt.scatter(x, y, color="r")
plt.title("Linear regression")
plt.ylabel("price")
plt.xlabel("sqft_living")
plt.plot(x, model.predict(x) )
```

```python
#multi linear regression
x=df[["sqft_living","grade"]]
y=df["price"].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.35,random_state=40)
print(x_train.shape)
print(y_train.shape)

model=LinearRegression()  
model.fit(x_train,y_train)  
predicted=model.predict(x_test) #testing our modelâ€™s performance
print("MSE", mean_squared_error(y_test,predicted))
print("R squared", metrics.r2_score(y_test,predicted))
```

```python
x=df["sqft_living"].values.reshape(-1,1)
y=df["price"].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.35,random_state=40)
poly=PolynomialFeatures(degree=2)
x_train_fit = poly.fit_transform(x_train)
model.fit(x_train_fit,y_train)
x_test_ = poly.fit_transform(x_test)
predicted = model.predict(x_test_)
print("MSE", mean_squared_error(y_test,predicted))
print("R squared", metrics.r2_score(y_test,predicted))
plt.scatter(x, y, color="r")
plt.title("Linear regression")
plt.ylabel("Salary")
plt.xlabel("Age")
plt.plot(x, model.predict(poly.fit_transform(x)), color="k") 
# multi linear regression fits more well than the poly and the linear in term of accuary
```
